merging player with mvp dataframe

```python
#removing * from players' names
players["Player"] = players["Player"].str.replace("*", "", regex = False)
```

```python
players.head(20)
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

```python
def single_row(df):
    if df. shape [0]==1:
        return df 
    else:
        row = df [df ["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1, :]["Tm"]
        return row
players = players.groupby (["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head(20)
```

*Combining Player and MVP Data

```python
combined = players.merge(mvps,how="outer", on=["Player","Year"])
```

```python
combined.head(5)
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined.head()
```

** Cleaning team data

```python
teams = pd.read_csv("teams.csv")
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(20)
```

```python
teams["Team"] =teams["Team"].str.replace("*","",regex = False)
```

```python
import csv

with open('nicknames.txt', 'r') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(",") for line in stripped if line)
    with open('nicknames.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerow(('title', 'intro'))
        writer.writerows(lines)
```

```python
nicknames = {}
with open ("nicknames.csv") as f:
    lines = f. readlines()
    for line in lines[1:]:
        abbrev , name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
stats = combined.merge (teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.apply(pd.to_numeric, errors = "ignore")
```
